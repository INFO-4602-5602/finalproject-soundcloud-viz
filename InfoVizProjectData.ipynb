```python
full = False
searchcollection = []
keyword = "pilots"
breakcount = 0
breaklimit = 10

for track in test.collection:
   # print(track.title, track.user["username"])
    searchcollection.append(track)

if hasattr(test,"next_href"):
    newlink = test.next_href
    try:
        while full == False:
            newlist = client.get(newlink)
            searchstring = ""
            for track in newlist.collection:
                if keyword in track.title.lower():
                    print(track.title," - " ,track.user["username"])
                    searchstring += track.title.lower()
                    searchcollection.append(track)
            if keyword not in searchstring:
                breakcount += 1
                print(" ---------- BREAK "+str(breakcount)+" ---------- ")
            if newlist.next_href == None or breakcount > breaklimit:
                print("full")
                full = True
            else:
                newlink = newlist.next_href
    except AttributeError:
        None

print(len(searchcollection))
```

The above code goes through the search string from test and if the keyword isn't anywhere in the request block (I use the title of the track) it stops.
